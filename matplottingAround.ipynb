```python
# importing modules
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
# arange(a, b, b) is a numpy method that creates an array of evenly c-spaced values from a to b, with only a included.
x = np.arange(0, math.pi*2, 0.5) # If q, w are within x, then w - q = 0.5. 0 belongs to x but not 2Ï€. 
print(x) # look at it, just to be clear
```

```python
y = np.sin(x) # strores each sine value from every element of x
print(y)
```

```python
plt.plot(x, y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
# If you want to plot a more precise sine wave, change the step of x:
x = np.arange(0, 2*math.pi, 0.05) # from 0.5 to 0.05
print(x)
y = np.sin(x)
print(y)
plt.plot(x, y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
# linspace(r, s, t) creates an array of t evenly spaced elements from r to s, r and s included.
# t is an optional parameter, its default value is 50.
a = np.linspace(-15, 35) 
print(a, len(a))
b = a**2
plt.plot(a, b, "r.")
plt.xlabel("X")
plt.ylabel("Y")
plt.title('parabola')
plt.show()
```

```python
a = np.linspace(-15, 15, 150) # t = 150 now
print(a, len(a))
b = a**2
plt.plot(a, b, "b-")
plt.xlabel("X")
plt.ylabel("Y")
plt.title('parabola')
plt.show()
```

```python
# This is good when you have to deal with just one simple plot object (plot for short). 
# But how to do things like a PRO? 

# Figures and axes

figure = plt.figure() # creates a figure, which is a window where to put plots. A figure has its own grid, so you 
# have to tell it exactly where to put your plots.

# a parabola
a = np.linspace(-15, 15, 150)
b = a**2

# Axes define a zone of the figure for displaying a plot.
# add_axes([x, y, a, b]) takes in a list of 4 numbers. These numbers goes from 0 to 1, both included.
# x, y denote the coordinates of the lowest point of the plot in this figure.
# a, b denote respectively width and heigth of the plot.
axes = figure.add_axes([0,0,0.8,0.5])  
axes.plot(a,b) # plots our parabola within these axes
axes.set_title("Parabola")
axes.set_xlabel('X')
axes.set_ylabel('Y')
plt.show()
```

```python
# How to plot more curves in just one plot:
x1 = [1, 16, 30, 42, 55, 68, 77, 88]
x2 = [1, 6, 12, 18, 28, 40, 52, 65]
x3 = [1, 9, 15, 21, 30, 37, 49, 63]
y = [1, 4, 9, 16, 25, 36, 49, 64]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # maximal dimensions

# Use color codes, marker codes and line code to style your curves.
l1 = ax.plot(x1,y,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x2,y,'go--') # dash line with green colour and circle marker
l3 = ax.plot(x3, y, 'mH:')

"""
legends(v, r) defines a legend for a plot.
It takes in a list of strings v - one string for each curve to plot - 
and either a string or numeric value that denotes its position. 
About r, these are the equivalent string-number pairs: 
Best = 0
upper right = 1
upper left = 2
lower left = 3
lower right = 4
Right = 5
Center left = 6
Center right = 7
lower center = 8
upper center = 9
Center = 10
"""
ax.legend(labels = ('TVs', 'Smartphones', 'Tablets'), loc = 'lower right') # legend placed at lower right
# or equivalently: ax.legend(labels = ('TVs', 'Smartphone', 'Tablets'), loc = 4)

ax.set_title("More curves in one plot: Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

```python
# How to display a plot inside another one:

x = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)

figure = plt.figure() # creating a new figure object

axes1 = figure.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes1.plot(x, y1, 'b')
axes1.set_title('sine')

axes2 = figure.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
axes2.plot(x, y2, 'r')
axes2.set_title("cosine")
plt.show()
```

```python
# Subplots

# How to display more plots at once:

# plt.subplots(h, j, k) creates a figure with a grid of 
# subplots, which are a special kind of plots, inside. 
# h is the number of rows, j is the number of columns.
# k is an index for enumerating these subplots: 
# it goes from 1 to k*j, from left to rigth and from up to down:
# So a (3,2) grid is indexed like this:
# [1]  [2]  [3]
# [4]  [5]  [6]


plt.subplot(3, 1, 1) # creates a grid of 3 rows and 1 column and then it points to subplot 1.
plt.plot([12, 34, 23, 66, 777, 120], 'cs') # creates a sequence envolving these numbers.

plt.subplot(3, 1, 2) # using the same figure, points to subplot 2.
plt.plot(range(12), 'mH') # creates a sequence with integers from 0 to 12.

# If you provide a single list or array to the plot() command, matplotlib assumes it is 
# a sequence of y values, and automatically generates the x values for you. 
# Since python ranges start with 0, the default x vector has the same length as y but starts with 0. 
# Hence the x data are [0, 12, 34, 23, 66, 777, 120] for the first case and range(12) for the second one.

a = [1, 9, 15, 21, 30, 37, 49, 63]
b = [1, 4, 9, 16, 25, 36, 49, 64]
plt.subplot(3, 1, 3) # using the same figure, points to subplot 3. You would write plt.subplot(313) too.
plt.plot(a, b, 'rD')
```
