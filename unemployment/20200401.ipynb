```python
import pandas as pd
from matplotlib import pyplot as plt

# taken from here: https://data.europa.eu/euodp/en/data/dataset/jrc-10113-rio_unemp_edlevel
data = pd.read_csv("/home/matteo/mypython/unemployment/RIO_UNEMP_EDLEVEL-data.csv")
metadata = pd.read_csv("/home/matteo/mypython/unemployment/RIO_UNEMP_EDLEVEL-meta.csv")
dataset = data[["GEO_DESC", "TIME_PERIOD", "SEX", "SEX_DESC", "AGE", "OBS_STATUS", "OBS_STATUS_DESC", "OBS_VALUE", "ISCED11"]]
dt = dataset[dataset.OBS_STATUS.notnull()]
dt
```

```python
# ED3: UPPER SECONDARY EDUCATION
# ED4: POST-SECONDARY NON-TERTIARY EDUCATION

# Unemployment % for ED3-4 educated population: Sweden

#ED3_4 Male case
swedenM = dt[dt.GEO_DESC == "Sweden"]
swedenM = swedenM[dt.SEX == "M"]
swedenM = swedenM[dt.OBS_STATUS != "bu"]
swedenM = swedenM[dt.OBS_STATUS != "u"]
swedenM = swedenM.sort_values(by = ["TIME_PERIOD"])
swedenM = swedenM[dt.ISCED11 == "ED3_4"] 
swedenM
```

```python
xm = swedenM.TIME_PERIOD
ym = swedenM.OBS_VALUE
plt.figure()
plt.plot(xm, ym)
plt.title("ED3_4")
plt.xlabel("years")
plt.ylabel("% unemployed males (15-74 aged)")
plt.show()
```

```python
#ED3_4 Female case
swedenF = dt[dt.GEO_DESC == "Sweden"]
swedenF = swedenF[dt.SEX == "F"]
swedenF = swedenF[dt.OBS_STATUS != "bu"]
swedenF = swedenF[dt.OBS_STATUS != "u"]
swedenF = swedenF.sort_values(by = ["TIME_PERIOD"])
swedenF = swedenF[dt.ISCED11 == "ED3_4"] 
swedenF
```

```python
xf = swedenF.TIME_PERIOD
yf = swedenF.OBS_VALUE
plt.figure()
plt.plot(xf, yf)
plt.title("ED3_4")
plt.xlabel("years")
plt.ylabel("% unemployed males (15-74 aged)")
plt.show()
```

```python
# T Total case
swedenT = dt[dt.GEO_DESC == "Sweden"]
swedenT = swedenT[dt.SEX == "T"]
swedenT = swedenT[dt.OBS_STATUS != "bu"]
swedenT = swedenT[dt.OBS_STATUS != "u"]
swedenT = swedenT.sort_values(by = ["TIME_PERIOD"])
swedenT = swedenT[dt.ISCED11 == "ED3_4"] 
swedenT
```

```python
xt = swedenT.TIME_PERIOD
yt = swedenT.OBS_VALUE
plt.figure()
plt.plot(xt, yt)
plt.title("ED3_4")
plt.xlabel("years")
plt.ylabel("% unemployed (15-74 aged)")
plt.show() 
```

```python
plt.figure()
plt.plot(xm, ym, "c:H")
plt.plot(xf, yf, "g-.D")
plt.plot(xt, yt, "m-s")
plt.title("ED3_4")
plt.xlabel("years")
plt.ylabel("% unemployed (15-74 aged)")
plt.legend(["male", "female", "total"])
plt.show()
```
